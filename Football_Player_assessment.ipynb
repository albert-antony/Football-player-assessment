```python
# Seperating categorical and numerical features

train_numeric = train [['row_id',
'scout_id',
'team',
'competitionId',
'player_general_var_1']]

train_cat = train.drop(['row_id',
'scout_id',
'team',
'competitionId',
'player_general_var_1'],axis=1)

train_cat.head()
```

#### Imputing missing values with median for numeric features and mode for categorical features

```python
for i in train_numeric.columns:
    train_numeric[i] = train_numeric[i].fillna(train_numeric[i].median(),axis=0)

train_numeric.isnull().sum()
```

```python
for j in train_cat.columns:
    train_cat[j] = train_cat[j].fillna(train_cat[j].mode()[0],axis=0)

train_cat.isnull().sum()

```

```python
# Concatenating both categorical and numerical features

train_new = pd.concat([train_numeric,train_cat],axis=1)

train_new.isnull().sum().to_csv("train_new.csv")
```

### Splitting the train data for evaluation

```python
###### train test split #####
# # split the data such that 20% of the data is set aside for testing
train_temp,test_temp=train_test_split(train_new,test_size=0.2,random_state=0)
print("Length of Original Data:-",len(train),"\nLength of Train Data:-",len(train_temp),"\nLength of Test Data:-",len(test_temp))

features=list(train_new.columns)
label='rating_num'


features.remove(label)

#print("\nFEATURES ARE:-",features)
#print("\nLABELS ARE:-",label)

# To check the data points in "Train" & "Test" datasets after split
X_train=train_temp[features]
y_train=train_temp[label]

X_test=test_temp[features]
y_test=test_temp[label]
```

# Modelling

## Base model -  Random Forest

```python
### Random Forest Regressor
model_lr1 = RandomForestRegressor(random_state=42)

#### Fit the model
model_lr1.fit(X_train,y_train) 

#### calculate algorithm training fit score
print("Train Data Set Score of 3rd Model is:-",model_lr1.score(X_train,y_train)) 

#### calculate algorithm test fit score
print("Test Data Set Score of 3d Model is:-",model_lr1.score(X_test,y_test)) 

print("----------------------------------------------------------------------------------------------------------------------")
```
